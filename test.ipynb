```python
from datetime import datetime, time, timedelta
import math

datetime_string = "6/7/2023 15:30:00"
time_delta_result = datetime.now() - datetime.strptime(datetime_string, "%m/%d/%Y %H:%M:%S")

total_seconds = time_delta_result.total_seconds()
# hours = int(total_seconds // 3600)
# minutes = int((total_seconds % 3600) // 60)
# seconds = int(total_seconds % 60)
# time_delta_str = "{:02d}:{:02d}:{:02d}".format(hours, minutes, seconds)
min = total_seconds // 60
print("Minutes: ", min)
print("Hours: ", min // 60)

this_time = datetime.now()
# last_start_time = datetime.now() - datetime.strptime(datetime_string, "%m/%d/%Y %H:%M:%S")
current_date = datetime.now().date()
start_of_day = datetime.combine(current_date, time.min)




# last_start_time = this_time - timedelta(minutes=1440)
time_passed = (this_time - start_of_day).total_seconds() // 60
data  = ((this_time - start_of_day).total_seconds() / 60 ) 

# print(timedelta(minutes=60), timedelta(minutes=120), timedelta(minutes=180), timedelta(minutes=1440), timedelta(minutes=7200))


# SCRIPT IS TRIGGERED EVERY 1 hour  
# Need specify reference date
# For 60 minutes --> each hour reset time to zero (the current minute is in the timeframe of 0 to 10)
# For 120 minutes --> check current hour % 2 == 0
# For 180 minutes --> check that current hour % 3 == 0
# For 240 minutes --> check current hour % 4 == 0
# For 480 minutes --> check that current hour % 8 == 0
# For 1440 minutes --> when the minutes value in the timeframe between 1400 - 1440
# For 7200 minutes --> Count the days passed starting from the refference date and check that days_passed % 5 == 0

def will_be_scraped_old(time_period):
    if datetime.now().minute in range(0,10):
        print("data")
    elif datetime.now().hour % (time_period//60) == 0:
        print("For 480 minutes/8 hours")
    elif datetime.now().hour % 4 == 0:
        print("For 240 minutes/4 hours")
    elif datetime.now().hour % 3 == 0:
        print("For 180 minutes/3 hours")
    elif datetime.now().hour % 2 == 0:
        print("For 120 minutes/2 hours")
    elif int(get_minutes_of_day()) in range(1400, 1440):
        print("For 1440 minutes/1 day")


def will_be_scraped(time_period, publish_date):
    time_period = int(time_period)

    one_hour_check = datetime.now().minute in range(0,10)
    other_hours_check = datetime.now().hour % (time_period//60)
    one_day_check = int(get_minutes_of_day()) in range(1380, 1440)

    time_frame = int(get_minutes_after_pulish(publish_date)) / 7200
    print(time_frame)
    decimal_part = round(time_frame % 1, 4)
    five_days_check = 0.9917 <= decimal_part or decimal_part <= 0.0083

    result = True

    if time_period == 60 and one_hour_check:
        print("For 60 minutes/1 hour")
    elif time_period in [120, 180, 240, 480] and other_hours_check == 0:
        print(f"For {time_period} minutes/{time_period//60} hours") 
    elif time_period == 1440 and one_day_check:
        print("For 1440 minutes/1 day")
    elif time_period == 7200 and  five_days_check:
        print("For 7200 minutes/5 days")
    else:
        result = False
    return result

def get_minutes_of_day():
    current_date = datetime.now().date()
    start_of_day = datetime.combine(current_date, time.min)
    time_passed = (datetime.now() - start_of_day).total_seconds() // 60
    return time_passed

def get_minutes_after_pulish(publish_date):
    time_delta_result = datetime.now() - datetime.strptime(publish_date, "%m/%d/%Y %H:%M:%S")
    total_seconds = time_delta_result.total_seconds()
    minutes_passed = total_seconds // 60
    return minutes_passed

# Example usage:
print("Minutes passed since the day started:", get_minutes_of_day())
print("Minutes passed:", get_minutes_after_pulish("7/20/2023 20:00:00"))
print('------------------')
print(will_be_scraped(120, "7/20/2023 20:00:00"))
```
